```python
import ollama
import random
import time


def gerar_dados_missao():

    temperatura = round(random.uniform(15.0, 110.0), 1)  # Pode passar de 80°C
    energia = random.randint(10, 100)                    # Pode ficar abaixo de 20%


    status_opcoes = ["ESTÁVEL", "OPERANDO COM RUÍDO", "FALHA CRÍTICA"]
    comunicacao = random.choice(status_opcoes)

    return temperatura, energia, comunicacao


def avaliar_regras_criticas(temp, energ, com):
    alertas_locais = []

    if temp > 80.0:
        alertas_locais.append("[ALERTA LOCAL] Temperatura crítica! Sistema de resfriamento ativado.")
    if energ < 20:
        alertas_locais.append("[ALERTA LOCAL] Bateria abaixo de 20%! Cortando energia de módulos secundários.")
    if com == "FALHA CRÍTICA":
        alertas_locais.append("[ALERTA LOCAL] Perda de sinal detectada! Reorientando antena primária.")

    return alertas_locais


def monitorar_missao():

    temperatura, energia, comunicacao = gerar_dados_missao()


    alertas_sistema = avaliar_regras_criticas(temperatura, energia, comunicacao)


    dados_formatados = f"Temperatura: {temperatura}°C | Energia: {energia}% | Comunicação: {comunicacao}"

    print("\n[INFO] Solicitando diagnóstico em tempo real para a IA...")

    resposta = ollama.chat(
        model="llama3.2:1b",
        messages=[
            {
                "role": "system",
                "content": (
                    "Você é o núcleo de Inteligência Artificial do Mission Control da FIAP. "
                    "Analise os dados operacionais recebidos e forneça um diagnóstico resumido "
                    "e uma recomendação técnica rápida para a tripulação. Seja direto e profissional."
                )
            },
            {
                "role": "user",
                "content": f"Dados atuais da nave: {dados_formatados}. Avalie o status atual."
            }
        ]
    )

    analise_ia = resposta["message"]["content"]


    print("=" * 55)
    print("         MISSION CONTROL AI — PAINEL OPERACIONAL")
    print("=" * 55)
    print(f" Temperatura dos Módulos : {temperatura}°C")
    print(f" Nível de Energia Solar  : {energia}%")
    print(f" Status da Comunicação   : {comunicacao}")
    print("-" * 55)


    if alertas_sistema:
        print(" SISTEMA DE DEFESA AUTOMÁTICO:")
        for alerta in alertas_sistema:
            print(f"   {alerta}")
    else:
        print(" SISTEMA DE DEFESA AUTOMÁTICO: Status Nominal (Seguro)")

    print("-" * 55)
    print(" ANÁLISE PREDITIVA DA IA:")
    print(f"{analise_ia.strip()}")
    print("=" * 55)


monitorar_missao()
```
